```python
N = 100
x_1 = 2 * np.random.random(N) - 1
x_2 = 2 * np.random.random(N) - 1

y_label = 0.2 * x_1 - 0.5 * x_2 #+ 0.2*(2*np.random.random(N)-1)
df = pd.DataFrame({"x1":x_1, "x2":x_2, "y":y_label})
df["ones"]=1
```

```python
X = df[["ones", "x1", "x2"]].values.T
Y = df.y.values.reshape(1, N)
```

```python
Nfeatures = np.shape(X)[0]
m = np.shape(X)[1]
print(Nfeatures, m)
```

```python
def model(theta, X):
  return theta.T@X

def costo(h, Y):
  J= (h-Y)**2
  return J.mean()

def update_params(theta, h, Y, learning_rate):
  """
  Gradiente descendente
  """
  grad = np.dot((h-Y), X.T).T
  theta = theta - learning_rate*grad
  return theta
```

```python
theta=np.random.random(3).reshape(3, 1)
learning_rate=0.01
Niter=100
J = np.zeros(Niter)

for i in range(0, Niter):
  h = model(theta, X)
  J[i] = costo(h, Y)
  theta = update_params(theta, h, Y, learning_rate)
```

```python
class MultilinearRegresion():
  def __init__(self, X, Y):
    self.X = X
    self.Y = Y
    Nfeatures = np.shape(X)[0]
    m = np.shape(X)[1]
    self.theta=np.random.random(Nfeatures).reshape(Nfeatures, 1)

  def model(self):
    self.h = self.theta.T@self.X

  def costo(self):
    self.J =  np.mean((self.h-self.Y)**2 )

  def update_params(self, learning_rate):
    """
    Gradiente descendente
    """
    grad = np.dot((self.h-self.Y), self.X.T).T
    self.theta = self.theta - learning_rate*grad

  def fit(self, learning_rate):
    for i in range(0, 100):
      self.model()
      self.costo()
      self.update_params(learning_rate)
```

```python
lr = MultilinearRegresion(X, Y)
```

```python
lr.fit(0.01)
```
