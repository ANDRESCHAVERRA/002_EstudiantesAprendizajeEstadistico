
9. Agrupar características similares y calcular su promedio.
   Para las siguientes características:



Usa expresiones regulares para identificar patrones comunes, como `radius1`, `radius2`,`radius3`, etc., y calcular los promedios. Por ejemplo:

```python
re.match(r'^[a-zA-Z_]+', "holamundo12341").group(0)
```

También puedes utilizar el método `startswith()` para buscar estos patrones en los nombres de las columnas.



````python
def agrupar(df, patron):
  grupo = {}
  for i in df.columns:
    match = re.match(patron,i)
    if match:
      l = match.group(0)
      if l not in grupo:
        grupo[l] = []
      grupo[l].append(i)
  return(grupo)

nombres = agrupar(df,r'^[a-zA-Z_]+')
nombres1 = list(nombres.keys())
nombres
````

````python
t = []
for i in range(1,11):
  columnas = nombres[nombres1[i]]
  datos = df[columnas]
  promedio = datos.mean(axis=1)
  t.append(promedio)
promedios =  pd.concat(t, axis=1)
nombre_columna = ['RadiusMean', 'TextureMean', 'PerimeterMean', 'AreaMean', 'SmoothnessMean', 'CompactnessMean', 'ConcavityMean', 'ConcavePointsMean', "SymmetryMean", "FractalDimensionMean"]
promedios.columns = nombre_columna
promedios
````
