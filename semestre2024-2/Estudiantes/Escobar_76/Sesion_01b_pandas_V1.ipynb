````python
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSHCOR8_Ha6TvBQwIcpjvJ0bzHYel1S8DXl4NHnMhVvdbibrgL_SP6rffuESpaJvPwLuUizXblQtHox/pub?output=csv"
df = pd.read_csv(url)
df
````

````python
titles = []
for i in range(len(df.columns)):
  if len(df.columns[i].split("_")) > 1:
      cambio = map(lambda x: x.capitalize(), df.columns[i].split("_"))
      titles.append("".join(cambio))
  else:
    titles.append(df.columns[i].capitalize())
titles
df.columns = titles
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset('penguins')
df
````

````python
df.dropna(axis = 0, inplace = True, ignore_index = True) # Limpiar el Dataset
df.info()
````

1. ¿Cuántas especies diferentes de pingüinos hay en el dataset?

````python
# este es el numero total de especies diferentes
numero_especies = len(df.species.unique())
print(f" El numero de especies es: {numero_especies}")
````

2. ¿Cuál es la longitud media del pico de los pingüinos de la especie Adelie?

````python
media_pico_Adelie = df.set_index('species').loc['Adelie'].bill_length_mm.mean()
print("La longitud media del pico de la especie Adelie: {media_pico_Adelie}")
````

3. ¿Cuál es la masa corporal promedio de los pingüinos de la especie Gentoo?

````python
masa_promedio_Getoo = df.set_index('species').loc['Gentoo'].body_mass_g.mean()
print(f"La masa promedio de la especie Gentoo: {masa_promedio_Getoo}")
````

4. ¿Cuál es la relación entre la longitud del pico y la masa corporal de los pingüinos?



````python
plt.scatter(df.bill_length_mm.values, df.body_mass_g.values )
plt.grid()
plt.xlabel('longitud de pico [mm]')
plt.ylabel('masa [g]')
````

````python
corr_pico_peso = df.bill_length_mm.corr(df.body_mass_g)
print(f"El pico y la masa corporal tienen una correlacion de: {corr_pico_peso}")
````

5. ¿Cuál es la isla con mayor número de pingüinos en el dataset?



````python
isla_max_pinguinos = df.island.value_counts(sort = True).reset_index()['island'].iloc(0)[0]
print(f"La isla con mayor numero de pingüinos es: {isla_max_pinguinos}")
````

6. ¿Cuál es la desviación estándar de la longitud de la aleta de los pingüinos de la especie Chinstrap?



````python
Chinstrap_pico_std = df.set_index('species').loc['Chinstrap'].flipper_length_mm.std()
print(f"la desviacion estándar de la longitud de la aleta de los pingüinos de la especie Chinstrap es: {Chinstrap_pico_std}")
````

7. ¿Qué pingüino tiene el pico más largo en el dataset y cuál es su especie?



````python
especie_pico_largo = df.set_index('bill_length_mm').loc[df.bill_length_mm.max()]['species']
print(f"La especie del pinguino pico más largo es: {especie_pico_largo}")
````

8. ¿Cuál es la proporción de machos y hembras en la especie Adelie?



````python
porcentaje_pinguinos_masculinos = (100 * df.sex.value_counts()['Male']) /len(df.sex)
porcentaje_pinguinos_femeninos = (100 * df.sex.value_counts()['Female']) /len(df.sex)
````

````python
Adelie_Nmachos = len(Adelie.set_index('sex').loc['Male'])
Adelie_Nfemeninos = len(Adelie.set_index('sex').loc['Female'])
Adelie_mediaflipperm = Adelie.set_index('sex').loc['Male'].flipper_length_mm.mean()
Adelie_mediaflipperf = Adelie.set_index('sex').loc['Female'].flipper_length_mm.mean()
````

````python
print(f"Para la especie Adelie hay {Adelie_Nmachos} machos y {Adelie_Nfemeninos} hembras")
print(f"para los machos tienen una longitud media de aleta {Adelie_mediaflipperm}")
print(f"para las feminas tienen una longitud media de aleta {Adelie_mediaflipperf}")
print(f"es claro entonces que la longitud promedio de los machos es mayor que la de las hembras.")
````

- **time**: Momento exacto en el que ocurrió el evento, en formato de fecha y hora.
- **latitude**: Latitud en grados decimales. Los valores negativos indican latitudes del hemisferio sur.
- **longitude**: Longitud en grados decimales. Los valores negativos corresponden a longitudes del hemisferio occidental.
- **depth**: Profundidad del evento en kilómetros.
- **mag**: Magnitud del terremoto.
- **magType**: Algoritmo o método utilizado para calcular la magnitud del evento.
- **nst**: Número total de estaciones sísmicas utilizadas para determinar la ubicación del terremoto.
- **gap**: Mayor brecha azimutal entre estaciones sísmicas adyacentes (en grados).
- **dmin**: Distancia horizontal desde el epicentro hasta la estación más cercana (en grados).
- **rms**: Raíz cuadrada media (RMS) del tiempo de viaje residual de las ondas sísmicas.
- **net**: Identificador de la red de datos que fue la fuente principal de información para este evento.
- **id**: Identificador único del evento.
- **updated**: Momento en que se actualizó la información del evento por última vez.
- **place**: Descripción textual de la región geográfica cercana al evento.
- **type**: Tipo de evento (por ejemplo, terremoto).
- **horizontalError**: Error estimado en la ubicación horizontal del epicentro.
- **depthError**: Error estimado en la profundidad del evento.
- **magError**: Error estimado en la magnitud del evento.
- **magNst**: Número de estaciones que reportaron la magnitud.
- **status**: Estado de revisión del evento (por ejemplo, revisado o automático).
- **locationSource**: Fuente de información sobre la ubicación del evento.
- **magSource**: Fuente de la magnitud informada del evento.
