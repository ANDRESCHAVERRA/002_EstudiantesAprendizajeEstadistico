7.

```python
df["DiagnosisisNumeric"] = df["Diagnosis"].map(lambda diag: 1 if diag == "M" else 0)
```

8.

```python
for col in df.columns.values:
    if col != "Diagnosis":
        df[col] = (df[col] - df[col].mean()) / df[col].std()
        
```

```python
df
```

9.

```python
for charac in ['RadiusMean', 'TextureMean', 'PerimeterMean', 'AreaMean', 'SmoothnessMean', 'CompactnessMean', 'ConcavityMean', 'ConcavePointsMean', "SymmetryMean", "FractalDimensionMean"]:
    start_of_word = charac.replace("Mean", "")
    sub_df = df.filter(regex=f'^{start_of_word}')
    df[charac] = sub_df.mean(axis=1)
```

```python
Q1 = df['RadiusMean'].quantile(0.25)
Q3 = df['RadiusMean'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['RadiusMean'] < (Q1 - 1.5 * IQR)) | (df['RadiusMean'] > (Q3 + 1.5 * IQR)))]
df
```
