````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset('penguins')
df.dropna(axis=0, inplace=True, ignore_index=True)

df.head()
````

## 1. ¿Cuántas especies diferentes de pingüinos hay en el dataset?

````python
print('1). Hay ' + str(len(df.species.unique())) + ' especies diferentes de pingüinos en el dataset, son:' + str(df.species.unique()) + '.')

sns.countplot(x='species', data=df)
plt.title('Cantidad de Pingüinos por Especie')
plt.xlabel('Especie')
plt.ylabel('Cantidad')
plt.show()
````

## 2. ¿Cuál es la longitud media del pico de los pingüinos de la especie Adelie?

````python
print('2). La longitud media del pico de los pingüinos de la especie Adelie es: ' + str(df[df.species == 'Adelie'].bill_length_mm.mean()))

adelie_df = df[df.species == 'Adelie']
x = range(len(adelie_df))
y = adelie_df.bill_length_mm

plt.scatter(x, y)
plt.title('Longitud del Pico de los Pingüinos Adelie')
plt.xlabel('Pingüino')
plt.ylabel('Longitud del Pico (mm)')

plt.axhline(y=df[df.species == 'Adelie'].bill_length_mm.mean(), color='red')

plt.show()
````

## 3. ¿Cuál es la masa corporal promedio de los pingüinos de la especie Gentoo?

````python
print('3). La masa corporal promedio de los pingüinos de la especie Gentoo es: ' + str(df[df.species == 'Gentoo'].body_mass_g.mean()))

gentoo_df = df[df.species == 'Gentoo']
x = range(len(gentoo_df))
y = gentoo_df.body_mass_g

plt.scatter(x, y)
plt.title('Masa Corporal de los Pingüinos Gentoo')
plt.xlabel('Pingüino')
plt.ylabel('Masa Corporal (g)')

plt.axhline(y=df[df.species == 'Gentoo'].body_mass_g.mean(), color='red')

plt.show()
````

## 4. ¿Cuál es la relación entre la longitud del pico y la masa corporal de los pingüinos?

Creamos una nueva columna para organizar los datos de la relacion entre la longitud del pico y la masa en gramos para cada uno de los pinguinos

````python
billBody = df.bill_length_mm/df.body_mass_g
df['bill_body_mass_rel'] = billBody
df.head()
````

````python
print('4). La relacion entre longitud del pico y masa corporal es: ' + str(billBody.mean()))


sns.scatterplot(x='bill_length_mm', y='body_mass_g', hue='species', data=df)

plt.title('Relación entre Longitud del Pico y Masa Corporal por Especie')
plt.xlabel('Longitud del Pico (mm)')
plt.ylabel('Masa Corporal (g)')
plt.legend(title='Especies')

plt.show()
````

## 5. ¿Cuál es la isla con mayor número de pingüinos en el dataset?

````python
countI = df.island.value_counts();
print('5). La isla con mayor numero de pinguinos en el dataset es: ' + (countI.idxmax(0)) + ' con ' + str(countI.max()) + ' pinguinos.')

sns.countplot(x='island', data=df)
plt.title('Cantidad de Pingüinos por Isla')
plt.ylabel('Cantidad')
plt.show()
````

## 6. ¿Cuál es la desviación estándar de la longitud de la aleta de los pingüinos de la especie Chinstrap?

````python
des = df[df.species == 'Chinstrap'].flipper_length_mm.std()
print('6). La desviacion estandar de la longitud de la aleta de los pingüinos de la especie Chinstrap es: ' + str(des))

Chinstrap_df = df[df.species == 'Chinstrap']
x = range(len(Chinstrap_df))
y = Chinstrap_df.flipper_length_mm

plt.scatter(x, y)
plt.title('Longitud del Pico de los Pingüinos Chinstrap')
plt.xlabel('Pingüino')
plt.ylabel('Longitud del Pico (mm)')

plt.axhline(y=df[df.species == 'Chinstrap'].flipper_length_mm.mean(), color='red')

plt.show()
````

Puesto que la desviacion estandar relaciona que tan alejados estan los datos de la media y en este caso, podemos ver como esta desviacion de 7.13 representada en los datos extremos en la grafica

## 7. ¿Qué pingüino tiene el pico más largo en el dataset y cuál es su especie?

````python
maxP = df.bill_length_mm.max()
print('El pico mas largo es de ' + str(maxP) + 'mm', 'y pertenece a la especie ' + str(df[df.bill_length_mm == maxP].species.unique()))
````

## 9. ¿Existe una correlación significativa entre la longitud del pico y la masa corporal en las diferentes especies de pingüinos?

````python
fig, axes = plt.subplots(1, 3 , figsize=(15, 5))


corrA = df[df.species == "Adelie"].bill_length_mm.corr(df[df.species == "Adelie"].body_mass_g)
sns.regplot(ax=axes[0], x='bill_length_mm', y='body_mass_g', data=df[df.species == 'Adelie'], color ='blue', line_kws={'color':'red'})
axes[0].set_title('Adelie\nCorrelación: ' + str(round(corrA , 2)))
axes[0].set_xlabel('Longitud del Pico (mm)')
axes[0].set_ylabel('Masa Corporal (g)')


corrG = df[df.species == "Gentoo"].bill_length_mm.corr(df[df.species == "Gentoo"].body_mass_g)
sns.regplot(ax=axes[1], x='bill_length_mm', y='body_mass_g', data=df[df.species == 'Gentoo'], color ='green', line_kws={'color':'red'})
axes[1].set_title('Gentoo\nCorrelación: ' + str(round(corrG , 2)))
axes[1].set_xlabel('Longitud del Pico (mm)')
axes[1].set_ylabel('Masa Corporal (g)')


corrC = df[df.species == 'Chinstrap'].bill_length_mm.corr(df[df.species == 'Chinstrap'].body_mass_g)
sns.regplot(ax=axes[2], x='bill_length_mm', y='body_mass_g', data=df[df.species == 'Chinstrap'], color ='orange', line_kws={'color':'red'})
axes[2].set_title('Chinstrap\nCorrelación:' + str(round(corrC , 2)))

axes[2].set_xlabel('Longitud del Pico (mm)')
axes[2].set_ylabel('Masa Corporal (g)')



plt.show()sns.regplot(ax=axes[2], x='bill_length_mm', y='body_mass_g', data=df[df.species == 'Chinstrap'], color ='orange', line_kws={'color':'red'})
````

Notemos que la correlacion entre longuitud de pico y masa corporal de la especie gento es la que mas se acerca a una correlacion lineal, sin embargo no podriamos afirmar que entre las especies haya una relacion significativa
