### 17. ¿Cuál es la diferencia en masa corporal promedio entre machos y hembras de la especie Chinstrap?

```python
especie_chinstrap = df[df['species'] == 'Chinstrap']

promedio_machos = especie_chinstrap[especie_chinstrap['sex'] == 'Male']['body_mass_g'].mean()
promedio_hembras = especie_chinstrap[especie_chinstrap['sex'] == 'Female']['body_mass_g'].mean()

diferencia = promedio_machos - promedio_hembras
print('El promedio de la masa corporal en las hembras de la especie Chinstrap es:',promedio_hembras)
print('El promedio de la masa corporal en los machos de la especie Chinstrap es:',promedio_machos)
print('La diferencia entre los promedios de machos y hembras de la especie Chinstrap es de: ',diferencia)
```
