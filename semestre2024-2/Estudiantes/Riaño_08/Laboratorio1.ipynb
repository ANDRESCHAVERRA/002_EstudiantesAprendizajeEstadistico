8. Normalizar cada columna respecto a su media y desviación estándar utilizando la fórmula:

````python
L = data2.columns.to_list()   
L.remove('Diagnosis')   #Remuevo Diagnosis para hacer el cálculo.
data3 = data2[L].apply(lambda x: (x-x.mean())/(x.std()))   #Realizo el cálculo para cada columna.
data3['Diagnosis'] = data2.Diagnosis   #Vuelvo a agregar Diagnosis.
data3.head()
````

12.  Determinar y eliminar los valores atípicos (`outliers`) en la columna `RadiusMean`.

Para identificarlos, construye un gráfico tipo `boxplot`:

```python
df.boxplot(column="RadiusMean", by='Diagnosis', sym='k.', figsize=(18,6))
```

Usa el rango intercuartílico (IQR) para definir los outliers. Los valores fuera del rango [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR] se consideran outliers. Ejemplo de cómo eliminar estos valores:

```python
Q1 = df['edad'].quantile(0.25)
Q3 = df['edad'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['edad'] < (Q1 - 1.5 * IQR)) | (df['edad'] > (Q3 + 1.5 * IQR)))]
```

Otra alternativa es utilizar el puntaje Z para detectar `outliers`:

```python 
df = df[(np.abs(stats.zscore(df['edad'])) < 3)]

````python
data3.boxplot(column="RadiusMean", by='Diagnosis', sym='k.', figsize=(8,5))
````

De la gráfica se puede ver que hay 6 outliers, 4 para la distribución de Diagnosis B y 2 para la distribución de Diagnosis M.

````python
#Outliers para diagnóstico B:

Diag_B = data3.RadiusMean[data3.Diagnosis=='B']

Q1_B = Diag_B.quantile(0.25)
Q3_B = Diag_B.quantile(0.75)
IQR_B = Q3_B - Q1_B

#Outliers para diagnóstico M:

Diag_M = data3.RadiusMean[data3.Diagnosis=='M']

Q1_M = Diag_M.quantile(0.25)
Q3_M = Diag_M.quantile(0.75)
IQR_M = Q3_M - Q1_M

#Filtrado para no tener en cuenta los outliers:

Filt_B = (~((Diag_B < (Q1_B - 1.5 * IQR_B)) | (Diag_B > (Q3_B + 1.5 * IQR_B))))   #Filtro para Diagnosis B.
Filt_M = (~((Diag_M < (Q1_M - 1.5 * IQR_M)) | (Diag_M > (Q3_M + 1.5 * IQR_M))))   #Filtro para Diagnosis M.

data4 = data3[pd.concat([Filt_B,Filt_M])]   #Junto los dos flitros y luego introduzco la máscara en data3.
````

Ahora verifiquemos que el filtro se haya realizado correctamente. Primero veamos cual era el tamaño original del dataframe.

````python
data3.shape
````

El tamaño es (569, 43). Ahora veamos el tamaño del nuevo dataframe filtrado. 

````python
data4.shape
````

El tamaño del nuevo dataframe es (563, 43). Es decir, 6 datos menos que el anterior. Por lo tanto, podemos inferir que el filtro se realizó correctamente, pues los outliers en las gráficas eran 6. Ahora apliquemos el otro filtro utilizando el puntaje Z.

````python
import numpy as np
import scipy.stats as stats

Filt_BZ = (np.abs(stats.zscore(Diag_B)) < 3)
Filt_MZ = (np.abs(stats.zscore(Diag_M)) < 3)

data5 = data3[pd.concat([Filt_BZ,Filt_MZ])]
````

Recordemos que el tamaño del dataframe original es (569, 43). Ahora veamos cuál es el tamaño del nuevo dataframe aplicando el filtro con el puntaje Z.

````python
data5.shape
````

Como podemos ver, el tamaño del nuevo dataframe filtrado usando el puntaje Z es (564, 43). Es decir, se eliminaron 5 outliers. Uno menos que usando el rango intercuartil.
