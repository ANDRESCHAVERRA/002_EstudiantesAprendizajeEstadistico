````python
import pandas as pd
import seaborn as sns
````

````python
df = sns.load_dataset('penguins')
df.dropna(axis=0, inplace=True, ignore_index=True)
df.head()
````

````python
sp_group = df.groupby('species')
sp_groups = [sp_group.get_group(sp) for sp in df['species'].unique()]
````

1. ¿Cuántas especies diferentes de pingüinos hay en el dataset?

````python
num_species = df['species'].nunique()
num_species
````

2. ¿Cuál es la longitud media del pico de los pingüinos de la especie Adelie?

````python
mean_peaks_adelie = float(sp_group['bill_length_mm'].mean()['Adelie'])
f"{mean_peaks_adelie:.2f}"
````

3. ¿Cuál es la masa corporal promedio de los pingüinos de la especie Gentoo?


````python
mean_mass_gentoo = float(sp_group['body_mass_g'].mean()['Gentoo'])
f"{mean_mass_gentoo:.2f}"
````

4. ¿Cuál es la relación entre la longitud del pico y la masa corporal de los pingüinos?


5. ¿Cuál es la isla con mayor número de pingüinos en el dataset?

````python
most_penguins = df.groupby('island')['species'].count().idxmax()
print(f'La isla con mas pinguinos es la isla {most_penguins}')
````

6. ¿Cuál es la desviación estándar de la longitud de la aleta de los pingüinos de la especie Chinstrap?

````python
std_chinstrap_fin_length = float(sp_group['flipper_length_mm'].std()['Chinstrap'])
f"La desviacion estandar para la aleta es {std_chinstrap_fin_length:.2f}"
````

7. ¿Qué pingüino tiene el pico más largo en el dataset y cuál es su especie?

````python
largest_peack = sp_group['bill_depth_mm'].max().idxmax()
print(f'La especie con el pico mas grande es {largest_peack}')
````

8. ¿Cuál es la proporción de machos y hembras en la especie Adelie?

````python
adelie_sex_count = sp_group.get_group('Adelie')['sex'].value_counts()
sex_prop_adelie = adelie_sex_count['Male']/adelie_sex_count['Female']
print(f'La proporcion de machos a hembras en la especie Adelie es {sex_prop_adelie} machos por cada hembra')
````

9. ¿Existe una correlación significativa entre la longitud del pico y la masa corporal en las diferentes especies de pingüinos?

````python
#correlaciones
corr = [group['flipper_length_mm'].corr(group['body_mass_g']) for group in sp_groups]

corrs = list(map(float, corr))

for i in range(num_species):
    print(f'La correlacion picos y masas en la especie {df["species"].unique()[i]} es {corrs[i]:.2f}')
````

Por los tanto se puede considerar una relacion lineal considerable entre la longitud de pico y la masa corporal en la especie Gentoo

11. ¿Cuál es la longitud mínima y máxima de la aleta entre las tres especies de pingüinos?

````python
min_fin_sp = map(float, [sp_group['flipper_length_mm'].get_group(sp).min() for sp in df['species'].unique()])
max_fin_sp = map(float, [sp_group['flipper_length_mm'].get_group(sp).max() for sp in df['species'].unique()])
vals_peack = list(zip(min_fin_sp, max_fin_sp))

for i in range(3):
    print(f'La especie {df["species"].unique()[i]} tiene un rango de aletas de {vals_peack[i][0]:.2f} a {vals_peack[i][1]:.2f} [mm]')
````

12. ¿Cuántos pingüinos tienen una masa corporal superior a 5000 gramos?

````python
mass_greater_5000 = df[df['body_mass_g'] > 5000]
print(f'Hay {mass_greater_5000.shape[0]} pinguinos con masa mayor a 5000g')
````

13. ¿Cómo varía la longitud del pico entre las diferentes especies de pingüinos en las distintas islas?

````python
mean_peack_adelie = float(sp_group['bill_length_mm'].mean()['Adelie'])
mean_peack_chinstrap = float(sp_group['bill_length_mm'].mean()['Chinstrap'])
mean_peack_gentoo = float(sp_group['bill_length_mm'].mean()['Gentoo'])


proportions = pow(np.array([mean_peack_adelie/mean_peack_chinstrap,
                mean_peack_adelie/mean_peack_gentoo,
                mean_peack_chinstrap/mean_peack_gentoo]), -1)

msg = [
    f'El pico de la especie Adelie es {proportions[0]:.2f} el tamaño que el de la especie Chinstrap',
    f'El pico de la especie Adelie es {proportions[1]:.2f} el tamaño que el de la especie Gentoo',
    f'El pico de la especie Chinstrap es {proportions[2]:.2f} el tamaño que el de la especie Gentoo'
]

for i in msg:
    print(i)
````

18. ¿Qué especie de pingüino tiene la longitud de aleta promedio más alta?

````python
sp_greater_mean_fin = [sp_group['flipper_length_mm'].get_group(sp).mean() for sp in df['species'].unique()]
sp_greater_mean_fin = list(map(float, sp_greater_mean_fin))
sp_greater_mean_fin_name = df["species"].unique()[sp_greater_mean_fin.index(max(sp_greater_mean_fin))]
print(f'La especie con la aleta mas larga en promedio es {sp_greater_mean_fin_name}')
````

20. ¿Cuál es la relación entre el sexo y la longitud de la aleta en los pingüinos de la especie Adelie?

````python
#correlaciones
corr = [group['sex'].map({'Male': 1, 'Female': 0}).corr(group['flipper_length_mm']) for group in sp_groups]

corrs = list(map(float, corr))
for i in range(num_species):
    print(f'La correlacion entre el sexo y la longitud de la aleta en la especie {df["species"].unique()[i]} es {corrs[i]:.2f}')
````

Por lo tanto, se tiene un vaga relacion lineal espercialmente en las especies Chinstrap y Gentoo, siendo la de Adelie menor
